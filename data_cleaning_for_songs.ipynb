```python
def prep(dosya):
  unwanted_words = ["Nakarat", "Bölüm", "Giriş", "Çıkış", "Ön-Nakarat", 
                    "Hook", "Chorus", "Verse", "Pre-Hook", "Intro", "Outro", 
                    "Bridge", "Instrumental", "Pre-Chorus", "Break", "Refrain", 
                    "Post-Chorus"]

  with open(dosya) as f:
    lines = f.readlines()

  for line in lines:
    for word in line.split():
      if word.replace("[", "").replace("]", "").replace("(", "").replace(")", "") in unwanted_words:
        try:
          lines.remove(line)
        except:
          line = line.replace(word, "")

  for line in lines:
    if "\n\n" in line:
      line.replace("\n\n", "\n")

  with open(dosya.replace(".txt", "") + "2" + ".txt", 'w') as f:
    for line in lines:
      f.write(line)
```

```python
prep("pop_veri_seti.txt")
prep("rock_veri_seti.txt")
prep("rap_veri_seti.txt")
```
